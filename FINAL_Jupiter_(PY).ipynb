```python
#2.1. Изучите распределение домов от наличия вида на набережную Постройте график Сделайте выводы
data = df['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

Круговая диаграмма (Pie chart) показывает, что большинство домов не имеют вид на набережную, в то время как домов с видом на набережную существенно меньше.

```python
#2.2. Изучите распределение этажей домов
data = df['floors'].value_counts()

plt.hist(df['floors'])
plt.title('Изучение распределения этажей домов')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
```

**Вывод:**

Гистограмма (Histogram) демонстрирует распределение домов по количеству этажей, позволяя увидеть, Что наибольшее количество домов - одноэтажные.

```python
#2.3. Изучите распределение состояния домов
data = df['condition'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

График демонстрирует распределение домов по разным уровням состояния, показывая, что большинство домов находится в удовлетворительном состоянии.

```python
#График зависимости стоимости от квадратуры жилой площади (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_house_data, x='sqft_living', y='price', alpha=0.7)
plt.xlabel('Квадратура жилой площади (sqft)')
plt.ylabel('Стоимость недвижимости')
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.show()
```

**Вывод:** Из графика видно, что есть положительная корреляция между квадратурой жилой площади и стоимостью недвижимости - чем больше площадь дома, тем выше его стоимость.

```python
# График распределения стоимости недвижимости по количеству комнат (box plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_house_data, x='bedrooms', y='price')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по количеству комнат')
plt.show()
```

**Вывод**: График показывает, что стоимость недвижимости варьируется в зависимости от количества комнат в доме. Дома с большим количеством комнат могут иметь более высокую стоимость, но есть также выбросы с низкой стоимостью для некоторых комбинаций количества комнат.

```python
#График распределения стоимости недвижимости по виду на набережную и количеству этажей (bar plot)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_house_data, x='waterfront', y='price', hue='floors')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по виду на набережную и этажам')
plt.legend(title='Количество этажей')
plt.show()
```

**Вывод:** Этот график позволяет сравнить среднюю стоимость недвижимости с видом на набережную и без него, а также учитывает влияние количества этажей. На графике видно, что дома с видом на набережную и несколькими этажами имеют более высокую среднюю стоимость.

```python
#График распределения стоимости недвижимости в зависимости от года постройки (box plot)
plt.figure(figsize=(20, 8))
sns.boxplot(data=df_house_data, x='yr_built', y='price')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по году постройки')
plt.xticks(rotation=90)
plt.show()
```

**Вывод:** График позволяет увидеть распределение стоимости недвижимости по годам постройки. Старые дома имеют разнообразные цены, но есть некоторые выбросы с высокой стоимостью. Дома, построенные недавно, имеют более высокую среднюю стоимость.
